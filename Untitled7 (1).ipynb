```python
import cv2
import numpy as np
import matplotlib.pyplot as plt

def hough_transform(image_path, max_lines=10, line_length=50):
    # Baca gambar
    image = cv2.imread(image_path, cv2.IMREAD_GRAYSCALE)

    # Terapkan edge detection (misalnya, menggunakan Canny)
    edges = cv2.Canny(image, 50, 150)

    # Terapkan Transformasi Hough
    lines = cv2.HoughLines(edges, 1, np.pi/180, threshold=100)

    # Gambar garis-garis yang terdeteksi ke dalam gambar asli
    result = cv2.cvtColor(edges, cv2.COLOR_GRAY2BGR)

    # Batasan jumlah garis yang digambar
    if lines is not None:
        lines = lines[:max_lines]

        for line in lines:
            rho, theta = line[0]
            a = np.cos(theta)
            b = np.sin(theta)
            x0 = a * rho
            y0 = b * rho

            # Sesuaikan panjang garis
            x1 = int(x0 + line_length * (-b))
            y1 = int(y0 + line_length * (a))
            x2 = int(x0 - line_length * (-b))
            y2 = int(y0 - line_length * (a))

            # Ganti warna garis menjadi hijau
            cv2.line(result, (x1, y1), (x2, y2), (0, 255, 0), 2)

    return result

# Gantilah dengan jalur gambar yang ingin Anda gunakan
image_path = "/content/corridor.jpg"

# Terapkan Transformasi Hough pada gambar dengan batasan garis dan panjang garis tertentu
hough_result = hough_transform(image_path, max_lines=5, line_length=30)

# Tampilkan gambar asli dan hasil Transformasi Hough
plt.subplot(121), plt.imshow(cv2.cvtColor(cv2.imread(image_path), cv2.COLOR_BGR2RGB)), plt.title('Gambar Asli')
plt.subplot(122), plt.imshow(cv2.cvtColor(hough_result, cv2.COLOR_BGR2RGB)), plt.title('Hasil Transformasi Hough')
plt.show()
```

```python
import cv2
import numpy as np
import matplotlib.pyplot as plt

def detect_circles(image_path, min_radius=10, max_radius=100):
    # Baca gambar
    image = cv2.imread(image_path, cv2.IMREAD_COLOR)

    # Konversi gambar ke skala abu-abu
    gray = cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)

    # Deteksi tepi menggunakan Canny
    edges = cv2.Canny(gray, 50, 150)

    # Lakukan transformasi Hough Lingkaran
    circles = cv2.HoughCircles(
        edges,
        cv2.HOUGH_GRADIENT,
        dp=1,
        minDist=20,
        param1=50,
        param2=30,
        minRadius=min_radius,
        maxRadius=max_radius
    )

    # Gambar lingkaran yang terdeteksi ke dalam gambar asli
    result = image.copy()

    if circles is not None:
        circles = np.uint16(np.around(circles))

        for i in circles[0, :]:
            # Gambar lingkaran
            cv2.circle(result, (i[0], i[1]), i[2], (0, 255, 0), 2)

    return result

# Gantilah dengan jalur gambar yang ingin Anda gunakan
image_path = "/content/bolla.jpg"

# Deteksi lingkaran pada gambar
circles_result = detect_circles(image_path)

# Tampilkan gambar asli dan hasil deteksi lingkaran
plt.subplot(121), plt.imshow(cv2.cvtColor(cv2.imread(image_path), cv2.COLOR_BGR2RGB)), plt.title('Gambar Asli')
plt.subplot(122), plt.imshow(cv2.cvtColor(circles_result, cv2.COLOR_BGR2RGB)), plt.title('Deteksi Lingkaran')
plt.show()
```
